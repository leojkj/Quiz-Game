```python
intro = input('do you want to play a game ').lower()

if intro == "yes":
    print("Let's play")
else:
    print("Good Bye")
    quit()

score = 0

question_1 = input('what is the first element in the periodic table ')

if question_1.lower() == 'hydrogen':
    print("Correct!")
    score += 1
else:
    print("Incorrect")

question_2 = input('Who does batman fight in The Dark Knight Rises ')

if question_2.lower() == "bane":
    print("Correct!")
    score += 1
else:
    print('Incorrect')

question_3 = input('What does a CPU stand for? ')

if question_3.lower() == 'computer processing unit':
    print("Correct!")
    score += 1
else:
    print('Incorrect')

question_4 = input('What is it called when one codes using only AI and the user does not type their own code ')

if question_4.lower() == 'vibe coding':
    print('Correct!')
    score += 1
else:
    print('Incorrect')

question_5 = input('Which U.S. city is nicknamed “The Windy City”? ')

if question_5.lower() == 'chicago':
    print('Correct')
    score += 1
else:
    print('Incorrect')

question_6 = input('Worth 2 points: Which sea creature has three hearts ')

if question_6.lower() == 'octopus':
    print('Correct!')
    score += 2
else:
    print('Incorrect')


print("You got " + str(score) + ' points!')
print(str(round((score/7)*100)) + "%")

print('Thank you for playing')
```
